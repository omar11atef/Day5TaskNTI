## Task 7 : 

```python
#Task 7 : # Function 
def analyze_scores(students_scores):
    if not students_scores:
        print("No student data.")
        return []

    total = 0
    for score in students_scores.values():
        total += score

    average = total / len(students_scores)
    print(f"Average score: {average:.2f}")

    highest_name = ""
    highest_score = -1
    for name, score in students_scores.items():
        if score > highest_score:
            highest_score = score
            highest_name = name
    print(f"Highest score: {highest_score} by {highest_name}")

    count_above_avg = 0
    below = []
    for name, score in students_scores.items():
        if score > average:
            count_above_avg += 1
        if score < 60:
            below.append(name)

    print(f"Number of students who scored above average: {count_above_avg}")
    return below

students = {
    "Ali": 80,
    "Sara": 45,
    "Ziad": 90,
    "Laila": 59
}

result = analyze_scores(students)
print("Students who scored below 60:", result)
```

## Task 8 : 

```python
def analyze_profits(products):
    total_profit = 0
    high_profit_products = []
    loss_products = []

    for name, data in products.items():
        cost = data['cost']
        revenue = data['revenue']
        profit = revenue - cost
        profit_percent = (profit / cost) * 100 if cost != 0 else 0

        print(f"{name}: Profit = {profit:.2f}")

        total_profit += profit

        if profit < 0:
            loss_products.append(name)

        if profit_percent > 30:
            high_profit_products.append(name)

    print(f"\nTotal Company Profit: {total_profit:.2f}")

    if loss_products:
        print("Products with loss:", ", ".join(loss_products))
    else:
        print("No product has a loss.")

    return high_profit_products

products = {
    "Laptop": {"cost": 700, "revenue": 950},
    "Phone": {"cost": 300, "revenue": 280},
    "Tablet": {"cost": 250, "revenue": 350},
    "Monitor": {"cost": 200, "revenue": 260}
}

result = analyze_profits(products)
print("Products with more than 30% profit:", result)
```

## Task 9 : 

```python

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def show_info(self):
        print(f"{self.name} - ${self.price} ({self.quantity} in stock)")

    def restock(self, amount):
        self.quantity += amount
        print(f"Added {amount} {self.name}(s). Now {self.quantity} in stock")

# Create products
apple = Product("MAC", 0.60, 200)
laptop = Product("Laptop", 999.99, 10)

# Test the products
print("=== Store Products ===")
apple.show_info()
laptop.show_info()

print("\n=== Restocking ===")
apple.restock(20)
laptop.restock(5)

print("\n=== Final Stock ===")
apple.show_info()
laptop.show_info()
```

## Task 10 : 

```python
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary
    
    def display_info(self):
        print(f"{self.name} ({self.position}): ${self.salary:,}")

class Company:
    def __init__(self, name):
        self.name = name
        self.employees = []
    
    def add_employee(self, employee):
        self.employees.append(employee)
        print(f"Added {employee.name} to {self.name}")
    
    def show_employees(self):
        print(f"\nEmployees at {self.name}:")
        for employee in self.employees:
            employee.display_info()
    
    def total_salary(self):
        total = sum(employee.salary for employee in self.employees)
        print(f"\nTotal monthly salary: ${total:,}")

tech_corp = Company("NTI")

OMAR = Employee("OMAR", "Developer", 7500)
Hamza = Employee("Hamza", "Manager", 9000)

tech_corp.add_employee(OMAR)
tech_corp.add_employee(Hamza)

tech_corp.show_employees()
tech_corp.total_salary()
```
