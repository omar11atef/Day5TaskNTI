```python
import numpy as np
import pandas as pd
```

```python
file_id = '1K-qffEKS6xNxtK-tM2jjFN9eHVj3H3ho'
url = f'https://drive.google.com/uc?id={file_id}'
```

```python
data = pd.read_csv(url)
data.head()
```

```python
data = data.drop('Poster_Link', axis=1)
```

```python
data.shape
```

```python
data.info()
```

```python
data['Released_Year'] = pd.to_numeric(data['Released_Year'], errors='coerce')
```

```python
data['Runtime'] = data['Runtime'].str.strip('min')
```

```python
data.head()
```

```python
data['Runtime'] = data['Runtime'].astype('float')
```

```python
data['Gross'] = data['Gross'].str.replace(',', '').astype(float)
```

```python
data.describe()
```

```python
data.describe(include='object')
```

```python
data['Certificate'].unique()
```

```python
data.isnull().sum()
```

```python
data.isnull().mean()*100
```

```python
data['Certificate'] = data['Certificate'].fillna('No Certificate')
```

```python
data['Meta_score'] = data['Meta_score'].fillna(data['Meta_score'].median())
```

```python
data.dropna(subset= ['Gross', 'Released_Year'], inplace=True)
```

```python
data.isnull().sum()
```

**Top 10 Directors by gross**

```python
pd.options.display.float_format = '{:,.0f}'.format
director_gross = data.groupby('Director')['Gross'].sum().sort_values(ascending=False).head(10)
director_gross
```

**Top 10 Star 1 by Gross**

```python
actor_gross = data.groupby('Star1')['Gross'].sum().sort_values(ascending=False).head(10)
actor_gross
```

**Top 3 years with the most movies (most active year)**

```python
data['Released_Year'].value_counts(sort=True).head(3)
```

**Average Runtime for drama movies (not useful insight but just for code)**

```python
drama_movies = data[data['Genre'].str.contains('Drama')]
average_runtime = drama_movies['Runtime'].mean()
average_runtime
```

**Average runtime for each decade (Does the runtime of movies changes when time passes?)**

```python
def extract_decade(x):
    first = int(x // 100)
    last = int((x % 100) // 10) * 10
    if last == 0:
        last = '00'

    return str(first) + str(last)
```

```python
data['decade']=data['Released_Year'].apply(extract_decade)
```

```python
data['decade'].unique()
```

```python
Runtime_by_decade = data.groupby('decade')['Runtime'].mean().sort_values(ascending=False)
Runtime_by_decade
```

We can see thar run time increased when time passed, but does it influence the gross?

```python
Runtime_by_decade = data.groupby('decade')[['Runtime', 'Gross']].mean().sort_values(by='Gross', ascending=False)
Runtime_by_decade
```

```python
Gross_by_decade = data.groupby('decade')['Gross'].mean().sort_values(ascending=False)
Gross_by_decade
```

**Top Actor by Total Gross**

```python
actors = pd.concat([
    data[['Star1', 'Gross']].rename(columns={'Star1': 'actor'}),
    data[['Star2', 'Gross']].rename(columns={'Star2': 'actor'}),
    data[['Star3', 'Gross']].rename(columns={'Star3': 'actor'}),
    data[['Star4', 'Gross']].rename(columns={'Star4': 'actor'}),
])

top_actor_gross = actors.groupby('actor')['Gross'].sum().sort_values(ascending=False).head(1)
top_actor_gross
```
