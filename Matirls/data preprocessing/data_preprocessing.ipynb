```python
#Outlier detection
#using IQR
#Chances of getting outliers are only in the 'Age' and 'Salary' columns.
```

```python
iqr=q3-q1
iqr
```

```python
upper=q3+1.5*iqr
upper
```

```python
#outlier rows
df[(df['Age']<lower) & (df['Age']>upper)]
#no outliers detected using IQR. But Age=0 is an outlier.
```
