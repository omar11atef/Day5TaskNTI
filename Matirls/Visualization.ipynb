```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(url)
data.head()
```

#Scatter plot

```python
plt.scatter(data['area'], data['price'], color = 'red', marker='x')
plt.show()
```

# Bar Plot

```python
furnishing_counts = data['furnishingstatus'].value_counts()
furnishing_counts
```

```python
furnishing_counts = data['furnishingstatus'].value_counts()
plt.bar(furnishing_counts.index, furnishing_counts.values)
plt.show()
```

```python
furnishing_counts = data['furnishingstatus'].value_counts()
colors = ['skyblue', 'green', 'red']

bars = plt.bar(furnishing_counts.index, furnishing_counts.values, color=colors)

plt.xlabel('Furnishing Status')
plt.ylabel('Number of Houses')
plt.title('Number of Houses by Furnishing Status')

plt.legend(bars, furnishing_counts.index, title='Furnishing Status')

plt.show()
```
