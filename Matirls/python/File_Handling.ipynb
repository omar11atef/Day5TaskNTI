Task 1 – Process a List of Numbers with Error Logging
Write a Python program that:

Has a predefined list of values:
data = [10, '5', 0, 'abc', 2]

Loops through each item and tries to divide 100 by the item.

For each item:

If it's valid and doesn't cause error → print the result, and save it to "results.txt" in the format:
"100 / value = result"

If the item is invalid (not a number or division by zero), catch the error and:

Print: "Error with value: value"

Save the message to "errors.txt" in the format:
"Failed to divide by value: reason"

At the end of the loop (after all items), use finally to write:
"Finished processing." at the end of both files.

```python
data = [10, '5', 0, 'abc', 2]

results_file = open("results.txt", "w")
errors_file = open("errors.txt", "w")

try:
    for item in data:
        try:
            value = float(item)
            result = 100 / value
        except ValueError as ve:
            print(f"Error with value: {item}")
            errors_file.write(f"Failed to divide by {item}: {ve}\n")
        except ZeroDivisionError as zde:
            print(f"Error with value: {item}")
            errors_file.write(f"Failed to divide by {item}: {zde}\n")
        else:
            print(f"100 / {value} = {result}")
            results_file.write(f"100 / {value} = {result}\n")
finally:
    results_file.write("Finished processing.\n")
    errors_file.write("Finished processing.\n")
    results_file.close()
    errors_file.close()
```

Task:
Create a Python program that allows users to:

Add a student’s name and grade to a CSV file called grades.csv.

Display all students and their grades from the file.
The program should keep running until the user decides to exit.

```python
import csv

print("Welcome to the Student Grades Program!")

while True:
    print("\n1. Add student")
    print("2. Show all students and grades")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        grade = input("Enter student grade: ")

        with open("grades.csv", "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([name, grade])
        print("Student added successfully!")

    elif choice == "2":
        try:
            with open("grades.csv", "r") as file:
                reader = csv.reader(file)
                print("\nStudents and Grades:")
                for row in reader:
                    if row:
                        print(f"{row[0]} - {row[1]}")
        except FileNotFoundError:
            print("No data found. Please add some students first.")

    elif choice == "3":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```

Task:
Create a Python program that allows the user to:

Add student names and numeric grades to a file called grades.csv.

Display the average grade of all students.

Display the name(s) of the student(s) who got the highest grade.
The program should repeat until the user chooses to exit.

```python
import csv

print("Welcome to the Student Grades System!")

while True:
    print("\n1. Add new student")
    print("2. Show average grade and top student(s)")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        grade = input("Enter student grade (number): ")

        with open("grades.csv", "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([name, grade])
        print("Student added successfully!")

    elif choice == "2":
        students = []

        try:
            with open("grades.csv", "r") as file:
                reader = csv.reader(file)
                for row in reader:
                    if row:
                        try:
                            name = row[0]
                            grade = float(row[1])
                            students.append((name, grade))
                        except ValueError:
                            continue
        except FileNotFoundError:
            print("No data found. Please add some students first.")
            continue

        if students:
            total = sum(g for _, g in students)
            avg = total / len(students)
            print(f"\nAverage grade: {avg:.2f}")

            highest = max(g for _, g in students)
            print("\nTop student(s):")
            for name, grade in students:
                if grade == highest:
                    print(f"{name} - {grade}")
        else:
            print("No valid data in file.")

    else:
        print("Exiting...")
        break
```
