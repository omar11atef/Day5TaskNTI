Task:
Write a function called analyze_scores that takes a dictionary of students and their scores. The function should:

Calculate and print the average score.

Print the name of the student with the highest score.

Count and print how many students scored above the average.

Return a list of the names of students who scored below 60.

```python
students = {
    "Ali": 50,
    "Sara": 75,
    "Mona": 90,
    "Omar": 100,
    "Laila": 65,
    "Tarek": 80
}
```

```python
def analyze_scores(students):
    total = 0
    highest_score = -1
    top_student = ""
    failed_students = []
    above_avg_count = 0

    for name, score in students.items():
        total += score
        if score > highest_score:
            highest_score = score
            top_student = name

    average = total / len(students)

    for name, score in students.items():
        if score > average:
            above_avg_count += 1
        if score < 60:
            failed_students.append(name)

    print(f"Average score is: {average:.2f}")
    print(f"Top student is: {top_student}")
    print(f"Number of students above average: {above_avg_count}")
    print(f"Students who failed: {failed_students}")

    return failed_students
```

Mini Project: Interactive Sales Analyzer
Build a Python program that simulates a business sales analysis tool.

Ask the user how many products they want to analyze.

Let the user input each product name and its monthly sales.

Store the data in a dictionary.

Pass the dictionary to a function called analyze_sales, which should:

Print the total sales.

Print the average sales.

Print the top-selling product.

Print the list of products below average.

Return a dictionary of percentage share of each product.

After calling the function, print the returned percentage shares.

```python
def analyze_sales(sales):
    total = sum(sales.values())
    average = total / len(sales)

    top_product = max(sales, key=sales.get)
    below_average = [product for product, value in sales.items() if value < average]

    product_shares = {}
    for product, value in sales.items():
        percent = (value / total) * 100
        product_shares[product] = round(percent, 2)

    print(f"\nTotal Sales: {total}")
    print(f"Average Sales: {average}")
    print(f"Top Product: {top_product}")
    print(f"Products below average: {below_average}")

    return product_shares

# Main program (user input)
sales = {}
num_products = int(input("How many products? "))

for _ in range(num_products):
    name = input("Enter product name: ")
    value = float(input(f"Enter sales for {name}: "))
    sales[name] = value

# Call the function
shares = analyze_sales(sales)

# Print the returned percentage shares
print("\nPercentage share of each product:")
for product, share in shares.items():
    print(f"{product}: {share}%")
```

Write a function called analyze_profits that takes a dictionary of product names and their cost and revenue in
The function should:

Calculate and print the profit for each product.

Print the total profit for the company.

Print the product(s) with negative profit (loss).

Return a list of products that made more than 30% profit.



```python
products = {
    "Laptop": {"cost": 7000, "revenue": 12000},
    "Mouse": {"cost": 500, "revenue": 1500},
    "Keyboard": {"cost": 1000, "revenue": 1200},
    "Monitor": {"cost": 4000, "revenue": 6000},
    "Headset": {"cost": 800, "revenue": 600}
}
```

```python
def analyze_profits(products):
    total_profit = 0
    loss_products = []
    high_profit_products = []

    for product, data in products.items():
        cost = data["cost"]
        revenue = data["revenue"]
        profit = revenue - cost
        total_profit += profit

        print(f"Profit for {product}: {profit}")

        if profit < 0:
            loss_products.append(product)

        if profit / cost > 0.3:
            high_profit_products.append(product)

    print(f"\nTotal Profit: {total_profit}")
    print(f"Products with loss: {loss_products}")

    return high_profit_products
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def display_info(self):
        print(f"Student name is {self.name}, and grade is {self.grade}")

    def check_pass(self):
        if self.grade >= 50:
            print(f"{self.name} has passed.")
        else:
            print(f"{self.name} has failed.")


student1 = Student("Mona", 90)
student2 = Student("Ali", 45)

student1.display_info()
student1.check_pass()

student2.display_info()
student2.check_pass()
```

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def display_info(self):
        print(f"Product: {self.name}")
        print(f"Price: {self.price}")
        print(f"Quantity: {self.quantity}")

    def calculate_total(self):
        return self.price * self.quantity

# Create product objects
product1 = Product("Laptop", 12000, 3)
product2 = Product("Mouse", 500, 10)

# Display info and total for product 1
product1.display_info()
print(f"Total price: {product1.calculate_total()}")

print()

# Display info and total for product 2
product2.display_info()
print(f"Total price: {product2.calculate_total()}")
```

Create two classes: Company and Employee.
The Company class should store a list of employees.
Add methods to:

Add a new employee

Display all employee information

Calculate the total salary for all employees
Then test your classes with sample data.

```python
class Employee:
    def __init__(self, name, base_salary, absent_days):
        self.name = name
        self.base_salary = base_salary
        self.absent_days = absent_days

    def display_info(self):
        print(f"Name: {self.name}, Base Salary: {self.base_salary}, Absent Days: {self.absent_days}")

    def calculate_salary(self):
        return self.base_salary - (self.absent_days * 100)

class Company:
    def __init__(self, name):
        self.name = name
        self.employees = []

    def add_employee(self, employee):
        self.employees.append(employee)

    def show_all_employees(self):
        print(f"\nEmployees in {self.name}:")
        for emp in self.employees:
            emp.display_info()

    def total_salaries(self):
        total = sum(emp.calculate_salary() for emp in self.employees)
        print(f"\nTotal Salaries Paid by {self.name}: {total}")

# Test the classes
emp1 = Employee("Ahmed", 5000, 1)
emp2 = Employee("Mona", 6000, 0)
emp3 = Employee("Omar", 4000, 3)

comp = Company("TechCorp")
comp.add_employee(emp1)
comp.add_employee(emp2)
comp.add_employee(emp3)

comp.show_all_employees()
comp.total_salaries()
```

You'll create two classes:

Book → represents a single book

Library → represents the whole library and stores a list of books
Each Book should have:

Title

Author

Year of publication

Your tasks:

Add a method to add a new book to the library

Add a method to search for a book by its title

Add a method to get all books written by a specific author (entered by the user).

If the author doesn't exist in the library, print a message like: "No books found by this author."



```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}, Year: {self.year}")


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def search_by_title(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                print("Book found:")
                book.display_info()
                return
        print("Book not found.")

    def get_books_by_author(self, author_name):
        found_books = [book for book in self.books if book.author.lower() == author_name.lower()]
        if found_books:
            print(f"Books by {author_name}:")
            for book in found_books:
                print(f"- {book.title} ({book.year})")
        else:
            print("No books found by this author.")


# Testing the classes:
library = Library()

book1 = Book("Harry Potter", "J.K. Rowling", 1997)
book2 = Book("Fantastic Beasts", "J.K. Rowling", 2001)
book3 = Book("1984", "George Orwell", 1949)

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.search_by_title("1984")
library.get_books_by_author("J.K. Rowling")
library.get_books_by_author("Unknown Author")
```
