# task

Using NumPy, build a small program to analyze sales data.

You are given a NumPy array of daily sales for a week (7 days).

Tasks:
Create a NumPy array that contains the sales data:
[1500, 2300, 1800, 0, 2700, 3000, 2000]

Print:

The total sales for the week

The average daily sales

The day(s) with the highest sales

The number of days with sales above 2000

The number of days with no sales

Replace all sales equal to 0 with the average of the week and print the new array.

```python
import numpy as np


sales = np.array([1500, 2300, 1800, 0, 2700, 3000, 2000])


total_sales = np.sum(sales)
average_sales = np.mean(sales)
max_sales = np.max(sales)
days_with_max = np.where(sales == max_sales)[0]
days_above_2000 = np.sum(sales > 2000)
days_with_zero = np.sum(sales == 0)


updated_sales = sales.copy()
updated_sales[updated_sales == 0] = average_sales


print("Original sales:", sales)
print("Total sales:", total_sales)
print("Average sales:", average_sales)
print("Highest sales:", max_sales)
print("Day(s) with highest sales (index):", days_with_max)
print("Number of days with sales > 2000:", days_above_2000)
print("Number of days with 0 sales:", days_with_zero)
print("Updated sales array:", updated_sales)
```

Analyze temperature data for 5 cities over 7 days using NumPy.

You are given a 2D NumPy array representing daily temperatures (in Celsius) for 5 cities over 7 days. Each row represents a city, and each column is a day.
Print:

The average temperature for each city.

The average temperature for each day.

The hottest day in each city.

The overall highest temperature and the city/day where it happened.

Number of days where any city's temperature was below 20°C.

```python
import numpy as np

temps = np.array([
    [30, 32, 33, 31, 29, 28, 30],
    [25, 26, 27, 28, 29, 30, 31],
    [20, 22, 23, 21, 24, 25, 26],
    [35, 34, 33, 32, 31, 30, 29],
    [15, 17, 18, 19, 20, 22, 23]
])

# 1. Average temperature for each city (row-wise)
avg_per_city = np.mean(temps, axis=1)

# 2. Average temperature for each day (column-wise)
avg_per_day = np.mean(temps, axis=0)

# 3. Hottest day in each city
hottest_per_city = np.max(temps, axis=1)

# 4. Overall highest temperature + location
overall_max = np.max(temps)
city_idx, day_idx = np.where(temps == overall_max)

# 5. Count of days with any city below 20
below_20 = temps < 20
days_with_below_20 = np.sum(np.any(below_20, axis=0))

# Print results
print("Average per city:", avg_per_city)
print("Average per day:", avg_per_day)
print("Hottest per city:", hottest_per_city)
print(f"Overall highest temp: {overall_max} at city {city_idx[0]}, day {day_idx[0]}")
print("Number of days with any city < 20°C:", days_with_below_20)
```
