else , finally

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ZeroDivisionError:
    print("You can't divide by zero.")
except ValueError:
    print("Please enter a number.")
else:
    print("Result is:", result)
finally:
    print("Program has finished running.")
```

**Task 1 – Password Length Validation with Error Handling**

Write a Python program that asks the user to enter a password.

If the password is 8 characters or more, print: "Password accepted."

If the password is less than 8 characters, print: "Password too short!"

If any unexpected error occurs, print: "Unexpected error occurred."

```python
try:
    password = input("Enter your password: ")
    if len(password) >= 8:
        print("Password accepted.")
    else:
        print("Password too short!")
except:
    print("Unexpected error occurred.")
```

**Task 2 – Validate and Process User Input List**
Write a Python program that:

Asks the user to input a list of numbers separated by commas (e.g., 10, 5, abc, 0).

Splits the input into a list.

Loops over each item and:

Tries to convert it to a number.

Divides 100 by the number.

For each item:

If the conversion or division fails:

Print: "Invalid value: item"

If successful:

Print: "100 / item = result"

After each iteration, use a finally block to print:
"Done with item: item"

```python
user_input = input("Enter numbers separated by commas: ")
items = user_input.split(',')

for item in items:
    item = item.strip()
    try:
        number = float(item)
        result = 100 / number
    except ValueError:
        print(f"Invalid value: {item}")
    except ZeroDivisionError:
        print(f"Invalid value: {item}")
    else:
        print(f"100 / {number} = {result}")
    finally:
        print(f"Done with item: {item}\n")
```

Task 3 – Retry Until Valid Grade is Entered
Write a Python program that:

Asks the user to input a grade (number between 0 and 100).

If the user enters:

A non-numeric value → print: "Invalid input. Please enter a number."

A number outside the range 0–100 → raise a ValueError manually with message: "Grade must be between 0 and 100."

Keep asking the user until a valid grade is entered.

If the grade is valid, print: "Grade accepted: [grade]"

Always print "Attempt finished." at the end of each try, using finally.



```python
while True:
    try:
        grade_input = input("Enter grade: ")
        grade = float(grade_input)

        if not (0 <= grade <= 100):
            raise ValueError("Grade must be between 0 and 100.")

    except ValueError as e:
        print(e)
    else:
        print(f"Grade accepted: {grade}")
        break
    finally:
        print("Attempt finished.\n")
```
